```python
import random
import time

# Función para cargar el banco de palabras desde un archivo de texto
def cargar_banco_palabras():
    with open("banco_palabras.txt", "r") as archivo:
        palabras = [line.strip().lower() for line in archivo]
    return palabras

# Función para jugar en el modo Normal
def modo_normal(banco_palabras):
    palabras_correctas = set(banco_palabras)
    puntos = 0
    tiempo_inicio = time.time()
    duracion_juego = 60  # 60 segundos

    respuestas_correctas = []

    while time.time() - tiempo_inicio < duracion_juego:
        palabra = input("Ingresa una palabra: ")
        palabra_minus = palabra.lower()
        if palabra_minus in palabras_correctas:
            puntos += 1
            respuestas_correctas.append(palabra_minus)

    print(f"\nTiempo agotado. Puntos: {puntos}")
    print('\nLas palabras correctas fueron:')
    imprimir_respuestas_ordenadas(respuestas_correctas)

# Función para jugar en el modo Por cantidad de caracteres
def modo_por_cantidad_de_caracteres(banco_palabras):
    caracteres = random.randint(4, 12)
    palabras_correctas = [palabra for palabra in banco_palabras if len(palabra) == caracteres]
    puntos = 0
    tiempo_inicio = time.time()
    duracion_juego = 60  # 60 segundos

    respuestas_correctas = []

    while time.time() - tiempo_inicio < duracion_juego:
        palabra = input(f"Ingresa una palabra de {caracteres} caracteres: ")
        palabra_minus = palabra.lower()
        if palabra_minus in palabras_correctas:
            puntos += 1
            respuestas_correctas.append(palabra_minus)

    print(f"\nTiempo agotado. Puntos: {puntos}")
    print('\nLas palabras correctas fueron:')
    imprimir_respuestas_ordenadas(respuestas_correctas)

# Función para jugar en el modo Por letra inicial
def modo_por_letra_inicial(banco_palabras):
    letra_inicial = random.choice('abcdefghijklmnopqrstuvwxyz')
    palabras_correctas = [palabra for palabra in banco_palabras if palabra.startswith(letra_inicial)]
    puntos = 0
    tiempo_inicio = time.time()
    duracion_juego = 60  # 60 segundos

    respuestas_correctas = []

    while time.time() - tiempo_inicio < duracion_juego:
        palabra = input(f"Ingresa una palabra que comience con '{letra_inicial}': ")
        palabra_minus = palabra.lower()
        if palabra_minus in palabras_correctas:
            puntos += 1
            respuestas_correctas.append(palabra_minus)

    print(f"\nTiempo agotado. Puntos: {puntos}")
    print('\nLas palabras correctas fueron:')
    imprimir_respuestas_ordenadas(respuestas_correctas)

#Ordenamiento de burbuja
def imprimir_respuestas_ordenadas(lista_correctas):
  for i in range(len(lista_correctas)-1, 0, -1):
    for j in range(i):
      if lista_correctas[j] > lista_correctas[j + 1]:
        lista_correctas[j], lista_correctas[j + 1] = lista_correctas[j + 1], lista_correctas[j]

  for palabra in lista_correctas:
    print(palabra)

def funcion_ayuda():
    ayuda = open("ayuda.txt", "r")
    a = ayuda.read()
    print(a)
    ayuda.close()

# Función principal del juego
def main():
    banco_palabras = cargar_banco_palabras()

    while True:
        print("\n** Menú **")
        print("1) Normal")
        print("2) Por cantidad de caracteres")
        print("3) Por letra inicial")
        print("4) Ayuda")
        print("5) Salir")
        opcion = input("Elige una opción: ")
        print('\n')

        if opcion == "1":
            modo_normal(banco_palabras)
        elif opcion == "2":
            modo_por_cantidad_de_caracteres(banco_palabras)
        elif opcion == "3":
            modo_por_letra_inicial(banco_palabras)
        elif opcion == "4":
            funcion_ayuda()
        elif opcion == "5":
            print("Gracias por jugar.")
            break
        else:
            print("Opción inválida")

main()
```
